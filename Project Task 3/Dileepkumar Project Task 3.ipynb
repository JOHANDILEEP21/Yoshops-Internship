```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import urllib.request
import re
from urllib.request import urlopen as uReq
from PIL import Image,ImageFont, ImageDraw
```

```python
product_name = []
product_name_without_image = []
product_url =[]
product_url_without_image =[]
product_image = []
product_without_image = []
product_description = []
tgttls = []
tglks = []
item_details=[]
baseurl = 'https://'

while True:
  try:
    print('\nEnter 1 put value link = \nEnter 2 put categories link= \nEnter 3 To find product page where Products Image missing')
    n = int(input())
    if n == 1:
      value_link = input('Enter Value link= ')
      print(value_link)
      # create excel file with  web url, Products name, Products Details, contact no and address columns.
      for x in range(1, 12):
        r = urllib.request.urlopen(baseurl+value_link).read()
        soup = BeautifulSoup(r,'lxml')
        productlist = soup.find('div', class_='collapse navbar-collapse').find_all('ul', class_='dropdown-menu')
        productlinks=[]
        for product in productlist:
          for link in product.find_all('a',class_=False, href=True, span=False):
            nam = link['href']
            if str(nam) in '/products':
              pass
            else:
              #productlinks.append(link['href'])
              productlinks.append(baseurl+value_link + link['href'])
      print(*productlinks)
    elif n == 2:
      product_name = []
      product_name_without_image = []
      product_url =[]
      product_url_without_image =[]
      product_image = []
      product_without_image = []
      product_description = []
      tgttls = []
      tglks = []
      item_details=[]

      cat_link = input('Enter category link: \n')
      #for i in range(len(productlinks)):
      print('You Were entered this Link: ', cat_link)
      url_1 = cat_link
      uClient_1 = uReq(url_1)
      page_html_1 = uClient_1.read()
      #uClient_1.close()
      if ((cat_link == 'https://yoshops.com/t/home--kitchen') or (cat_link == 'https://yoshops.com/t/electronics')):
        r = urllib.request.urlopen(cat_link).read()
        soup = BeautifulSoup(r,'lxml')
        url = cat_link
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
        req = requests.get(url, headers = headers).text
        soup = BeautifulSoup(req, 'html.parser')
        containers = soup.findAll("div", {"class":"col-sm-3 col-xs-6"})
        for container in containers:
          #print(container.div.a.span.text)
          product_name.append(container.div.a.span.text)

          product_link = baseurl + value_link + container.div.a.span.get('href')
          product_url.append(product_link)
          
          product_image_link = container.div.a.div.div.img['src']
          product_image.append(product_image_link)
          
          if (product_image_link == "//onlinestore.wsimg.com/assets/noimage/product-5fec99477aebb10bac85d82665ec1497de4536cda3279e59089555c45cf589fa.png"):
            product_name_without_image.append(container.div.a.span.text)
            product_url_without_image.append(product_url)
            product_without_image.append(product_image_link)
            url_2 = baseurl + value_link + container.div.a.span.get('href')
            uClient_2 = uReq(url_2)
            page_html_2 = uClient_2.read()
            soup = BeautifulSoup(page_html_2,"html.parser")
            product_details = soup.findAll("div", {"class":"single-product-description"})
            item_details=[]
            for details in product_details:
                item_details.append(details.text.strip('\n\n ').strip('\n\n\n\n\n ').strip('Buy ').strip())
            product_description.append(*item_details)
      else:
        for x in range(1):
          r = urllib.request.urlopen(cat_link).read()
          soup = BeautifulSoup(r,'lxml')
          page_num = soup.find('ul', class_='pagination pull-right').find_all('li', class_='num')
          pages = list(range(1,len(page_num)+1))
          for page in pages:
            url = cat_link+'?page={}'
            headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
            req = requests.get(url.format(page), headers = headers).text
            soup = BeautifulSoup(req, 'html.parser')
            containers = soup.findAll("div", {"class":"col-sm-3 col-xs-6"})

            for container in containers:
              #print(container.div.a.span.text)
              product_name.append(container.div.a.span.text)

              product_link = baseurl + value_link + container.div.a.span.get('href')
              product_url.append(product_link)
              
              product_image_link = container.div.a.div.div.img['src']
              product_image.append(product_image_link)
              
              if (product_image_link == "//onlinestore.wsimg.com/assets/noimage/product-5fec99477aebb10bac85d82665ec1497de4536cda3279e59089555c45cf589fa.png"):
                product_name_without_image.append(container.div.a.span.text)
                product_url_without_image.append(product_url)
                product_without_image.append(product_image_link)
                url_2 = baseurl + value_link + container.div.a.span.get('href')
                uClient_2 = uReq(url_2)
                page_html_2 = uClient_2.read()
                soup = BeautifulSoup(page_html_2,"html.parser")
                product_details = soup.findAll("div", {"class":"single-product-description"})
                item_details=[]
                for details in product_details:
                    item_details.append(details.text.strip('\n\n ').strip('\n\n\n\n\n ').strip('Buy ').strip())
                product_description.append(*item_details)
      
      missing_product_image = pd.DataFrame(list(zip(product_url_without_image, product_name_without_image, product_description)),
                   columns =['Web URL', 'Product Name', 'Product Details'])
      print(missing_product_image)
      missing_product_image.to_excel('/content/drive/MyDrive/YoShops Internship/Project Task 3/Product page Automation Validation.xlsx', sheet_name='Missing Products Images', index=False)
    elif n == 3:
      product_name = []
      product_name_without_image = []
      product_url =[]
      product_url_without_image =[]
      product_image = []
      product_without_image = []
      product_description = []
      tgttls = []
      tglks = []
      item_details=[]

      value_link = input('Enter Value link= ')
      print(value_link)
      for x in range(1, 12):
        r = urllib.request.urlopen(baseurl+value_link).read()
        soup = BeautifulSoup(r,'lxml')
        productlist = soup.find('div', class_='collapse navbar-collapse').find_all('ul', class_='dropdown-menu')
        productlinks=[]
        for product in productlist:
          for link in product.find_all('a',class_=False, href=True, span=False):
            nam = link['href']
            if str(nam) in '/products':
              pass
            else:
              #productlinks.append(link['href'])
              productlinks.append(baseurl+value_link + link['href'])
      print(*productlinks)
      for i in range(0, len(productlinks)):
        print('Working with this Link: ', productlinks[i])
        url_1 = productlinks[i]
        uClient_1 = uReq(url_1)
        page_html_1 = uClient_1.read()
        #uClient_1.close()
        if ((url_1 == 'https://yoshops.com/t/home--kitchen') or (url_1 == 'https://yoshops.com/t/electronics')):
            r = urllib.request.urlopen(url_1).read()
            soup = BeautifulSoup(r,'lxml')
            url = url_1
            headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
            req = requests.get(url_1, headers = headers).text
            soup = BeautifulSoup(req, 'html.parser')
            containers = soup.findAll("div", {"class":"col-sm-3 col-xs-6"})
            for container in containers:
              #print(container.div.a.span.text)
              product_name.append(container.div.a.span.text)

              product_link = baseurl + value_link + container.div.a.span.get('href')
              product_url.append(product_link)
              
              product_image_link = container.div.a.div.div.img['src']
              product_image.append(product_image_link)
              
              if (product_image_link == "//onlinestore.wsimg.com/assets/noimage/product-5fec99477aebb10bac85d82665ec1497de4536cda3279e59089555c45cf589fa.png"):
                product_name_without_image.append(container.div.a.span.text)
                product_url_without_image.append(product_url)
                product_without_image.append(product_image_link)
                url_2 = baseurl + value_link + container.div.a.span.get('href')
                uClient_2 = uReq(url_2)
                page_html_2 = uClient_2.read()
                soup = BeautifulSoup(page_html_2,"html.parser")
                product_details = soup.findAll("div", {"class":"single-product-description"})
                for details in product_details:
                    item_details.append(details.text.strip('\n\n ').strip('\n\n\n\n\n ').strip('Buy ').strip())
                product_description.append(*item_details)
        else:
          for x in range(1):
            r = urllib.request.urlopen(productlinks[i]).read()
            soup = BeautifulSoup(r,'lxml')
            page_num = soup.find('ul', class_='pagination pull-right').find_all('li', class_='num')
            pages = list(range(1,len(page_num)+1))
            for page in pages:
              url = productlinks[i]+'?page={}'
              headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
              req = requests.get(url.format(page), headers = headers).text
              soup = BeautifulSoup(req, 'html.parser')
              containers = soup.findAll("div", {"class":"col-sm-3 col-xs-6"})

              for container in containers:
                #print(container.div.a.span.text)
                product_name.append(container.div.a.span.text)

                product_link = baseurl + value_link + container.div.a.span.get('href')
                product_url.append(product_link)
                
                product_image_link = container.div.a.div.div.img['src']
                product_image.append(product_image_link)
                
                if (product_image_link == "//onlinestore.wsimg.com/assets/noimage/product-5fec99477aebb10bac85d82665ec1497de4536cda3279e59089555c45cf589fa.png"):
                  product_name_without_image.append(container.div.a.span.text)
                  product_url_without_image.append(product_link)
                  product_without_image.append(product_image_link)
                  url_2 = baseurl + value_link + container.div.a.span.get('href')
                  uClient_2 = uReq(url_2)
                  page_html_2 = uClient_2.read()
                  soup = BeautifulSoup(page_html_2,"html.parser")
                  product_details = soup.findAll("div", {"class":"single-product-description"})
                  item_details=[]
                  for details in product_details:
                      item_details.append(details.text.strip('\n\n ').strip('\n').strip('\n\n\n\n\n ').strip('Buy ').strip())
                  product_description.append(*item_details)

      missing_product_image = pd.DataFrame(list(zip(product_url_without_image, product_name_without_image, product_description)),
                  columns =['Web URL', 'Product Name', 'Product Details'])
      print(missing_product_image)
      missing_product_image.to_excel('/content/drive/MyDrive/YoShops Internship/Project Task 3/Page Automation Validation 1.xlsx', sheet_name='Total Image Missing Products', index=False)
    else:
      break
  except Exception as e:
    print(e)
```
