## 3. Format data in Excel sheet

```python
# Load the workbook
wbk = openpyxl.load_workbook('Task-1.xlsx')

# Get the worksheet by name
wst = wbk['Sheet']

# Format the header row
header = wst[1]
for cell in header:
    cell.font = openpyxl.styles.Font(bold=True, italic=True)
    cell.alignment = openpyxl.styles.Alignment(horizontal='center')

# Format the data rows
for row in wst.iter_rows(min_row=2):
    for cell in row:
        cell.number_format = '#,##0.00'

# Save the workbook
wbk.save('Task-1_formatted.xlsx')

wb1 = xlrd.open_workbook('Task-1_formatted.xlsx')
wb1
```

## 5.Python program for Extract mobile no from PDF, Json,XML and MS word file and save into MS excel

```python
pip install docx2txt
```

```python
path=r'/content/drive/MyDrive/YoShops Internship/Task - 1/1657173630381_5042579018065678/chats/2022'
def get_file(path):
    files=[]
    file_list=[]
    #r=root, d=directory, f=files
    for r,d,f in os.walk(path):
        for file in f:
            if '.json' in file:
                files.append(os.path.join(r,file))
    for f in files:
        
        file_list.append(f)
    return file_list
list_of_paths=get_file(path)
```

```python
mob1=[]
mob2=[]
for i in list_of_paths:
    with open(i,'r') as f:
        docs=json.loads(f.read())
    number1=''
    dic1 = docs['messages'][0]
    dic2 = docs['messages'][1]
    for i in dic1['msg']:
        if i.isdigit():
            number1+=i
    mob1.append(number1)
    number2=''
    for j in dic2['msg']:
        if j.isdigit():
            number2+=j
    mob2.append(number2)
df = pd.DataFrame.from_dict({'Yoshops':mob1,'Customers':mob2})
df.to_excel('/content/Demo.xlsx', header=True, index=False)
mob1
```
